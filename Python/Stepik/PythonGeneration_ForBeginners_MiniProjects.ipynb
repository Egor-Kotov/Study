```python
"""Угадайка чисел"""

from random import randint


def guess_number_game():
    is_play = True

    while is_play:

        print("\nДобро пожаловать в числовую угадайку.\n")
        number = input(
            "В каких пределах Вам загадать число?\nОт 1 до N включительно.  Введите N: "
        )

        while not number.isdigit():
            print("\nВведите, пожалуйста, число от 1!")
            number = input(
                "В каких пределах вам загадать число?\nОт 1 до N включительно.  Введите N: "
            )

        N = randint(1, int(number))
        guess = None
        attempts = 0
        continue_play = ""

        print("\n...Загадываю...\n")

        while guess != N:

            guess = input("\nУгадывайте! N = ")

            while not guess.isdigit() and 0 > int(guess) or int(guess) > int(number):
                print(f"\nВведите, пожалуйста, число от 1 до {number}!")
                guess = input("\nУгадывайте! N = ")

            guess = int(guess)
            attempts += 1

            if guess > N:
                print("\nВаше число больше загаданного, попробуйте еще разок")
            elif guess < N:
                print("\nВаше число меньше загаданного, попробуйте еще разок")
            else:
                print(f"\n\t\tВы угадали, поздравляем! Количество попыток: {attempts}")

        continue_play = input(
            "\nХотите, ещё партейку??? Введите 'д' = (да) или 'н' = (нет): "
        )

        while continue_play.lower() not in ("д", "да", "н", "нет"):
            continue_play = input("Введите, пожалуйста: д или н - ")

        if continue_play in ("н", "нет"):
            is_play = False


guess_number_game()
```

```python
"""Висильница"""

import requests, random
from bs4 import BeautifulSoup
from IPython.display import Image


"""Создаю список слов (1000 самых употребляемых существительных/глаголов/прилогательрных) с сайта http://dict.ruslang.ru/"""

urls = ('http://dict.ruslang.ru/freq.php?act=show&dic=freq_adj&title=%D7%E0%F1%F2%EE%F2%ED%FB%E9%20%F1%EF%E8%F1%EE%EA%20%E8%EC%E5%ED%20%EF%F0%E8%EB%E0%E3%E0%F2%E5%EB%FC%ED%FB%F5',
       'http://dict.ruslang.ru/freq.php?act=show&dic=freq_v&title=%D7%E0%F1%F2%EE%F2%ED%FB%E9%20%F1%EF%E8%F1%EE%EA%20%E3%EB%E0%E3%EE%EB%EE%E2',
       'http://dict.ruslang.ru/freq.php?act=show&dic=freq_s&title=%D7%E0%F1%F2%EE%F2%ED%FB%E9%20%F1%EF%E8%F1%EE%EA%20%E8%EC%E5%ED%20%F1%F3%F9%E5%F1%F2%E2%E8%F2%E5%EB%FC%ED%FB%F5',)

slova = []

for u in urls:
    words = []
    
    url = u
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    
    for string in soup.stripped_strings:
        words.append(string)
    
    slova += words[12::4][:-2]
    
# чистим от длиных и иностранных слов,  встряхиваем лист 
slova = [w for w in slova if len(w) < 18 and w.isalpha() and w[0].lower() not in 'abcdefghijklmnopqrstuvwxyz']
random.shuffle(slova)


slova = tuple(slova)


class Hangman:

#     Нужные иллюстрации и тд
    def __init__(self):
        self.letters_idx = []
        self.image = ("https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/start.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/one_mist.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/two_mist.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/three_mist.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/four_mist.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/five_mist.jpg", "https://raw.githubusercontent.com/Egor-Kotov/Python/main/image/Hangman/end.jpg")
    
#         Вход в игру и и выбор количества открытых букв 
    def start(self):
        start = input("Сыграем в Виселицу? (Выбери да или нет) ")
        if start == 'да':
#           Выбираем сколько букв открыто в начале
            self.letters = int(input("\nСколько букв известно в начале?\n0, 1 или 2? "))
            return True
#          Если игрок отказывается играть, возращаем Файлс для закрытия проги
        else:
            print("Byu-byu")
            return False
 
        
    def choose_word(self):
#        Выбираем слово для игры
        self.hidden_word = slova[random.randint(0, len(slova)-1)]
#         Если слово короче, чем количесвто открытых в начале букв, выбираем другое слово
        if len(self.hidden_word) <= self.letters:
            return choose_word()

        
    def l_idx(self, t = 0):
#         Генерируем индексы куда вставить открытые буквы
        if t == 0:
#         при t = 0 генерируется изначальный-рандомный список индексов для открытых букв
            while len(self.letters_idx) < self.letters:
                i = random.randint(0, len(self.hidden_word)-1)
                if i not in self.letters_idx:
                    self.letters_idx.append(i)
#         при t > 0 генерируется список открытых букы + индексом угаданных букв
        else:
#           переменую для выявления изминений
            self.change = False
            for i in range(len(self.hidden_word)):
#           Проходимся по всем элиментам, если его ирндекса нет в списке открытых элиментов
#           и этот элемент guess, добовляем его индекс в список
                if i not in self.letters_idx and self.guess == self.hidden_word[i]:
                    self.letters_idx.append(i)
                    self.change = True
#           когда прошла итеррация проверяем были ли изминнения и если были возращаем Тру
            if self.change is True:
                return True

            
    def rebus(self):       
#          Генерируем ребус основываясь на индексах открытых букв, какждый раз с нуля
        self.rebus_word = ""
        for i in range(len(self.hidden_word)):
            if i not in self.letters_idx:
                self.rebus_word = "%s%s" % (self.rebus_word, "&")
            else:
                self.rebus_word = "%s%s" % (self.rebus_word, self.hidden_word[i])

                
#            Игровой процесс   
    def play(self):
#     Создаем количество попыток, знаем что их у нас 6
        self.trying = 1
        print("Let's go!")
        display(Image(url=self.image[0]))
#      Запускаем цикл, который будет спрашивать букву пока, есть попытки или слово не угадано
        while self.trying < 7 and self.hidden_word != self.rebus_word:
#      Спаршиваем букву
            self.guess = input(f'Какая буква? {self.rebus_word} ? ')
#      Запускаем метод который, првоерит и добавит индекс буквы если она есть в список открытых букв
            if self.l_idx(self.trying):
#      Если есть изминения собираем слово-ребус заново
                self.rebus()
            else:
#     Если ошибка, покажeм картинку, и уменьши количесвто оставшихся попыток
                print("\n\tУВЫ(((")
                display(Image(url=self.image[self.trying]))
                self.trying += 1
                if self.trying == 7:
                    print(f"\n\t{self.hidden_word}")
#       После выхода из цикла, если слово угадоно объявляем победу
        if self.hidden_word == self.rebus_word:
            print(f"\n\n\n{self.hidden_word}\n\tWin!!!")
        

#  Собираем игру в один метод   
    def game(self):
        if self.start():
            self.choose_word()
            self.l_idx()
            self.rebus()
            self.play()
            
    
h = Hangman()        
h.game()
```
